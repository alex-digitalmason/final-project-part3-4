```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/greg/Downloads/signup_follow_projects_success_scrubbed.csv')
```

```python
def convert_to_list(proj_id_string):
    proj_id_string = proj_id_string[1:-1]
    proj_id_string_list = proj_id_string.split(', ')
    proj_id_int_list = list(map(int, proj_id_string_list))
    return proj_id_int_list

df.loc[:, 'project_ids'] = df.loc[:, 'project_ids'].apply(convert_to_list)
```

```python
split_data = {
    'action': [],
    'project_ids': [],
    'user_id': [],
}

for _, row in df.iterrows():
    for proj_id in row.loc['project_ids']:
        split_data['action'].append(row.loc['action'])
        split_data['project_ids'].append(proj_id)
        split_data['user_id'].append(row.loc['user_id'])
```

```python
split_data
```

```python
pd.DataFrame(split_data)
```

```python
pd.DataFrame(split_data).to_csv('signup_follow_projects_success_scrubbed_split.csv')
```
